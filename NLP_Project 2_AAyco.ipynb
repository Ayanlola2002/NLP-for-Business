3. Do topic modeling on the text after due preprocessing, vectorization etc.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# stop word removal using nltk
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
for i in data:
    print(i)
```

```python
# set of stop words
stop_words = set(stopwords.words('english')) 
word_tokens=[]
for i in data:
    tokenized = word_tokenize(i)
    word_tokens.append(tokenized)

for w in word_tokens:
    print(w)
```

```python
filtered_sentence = [] 
for i in word_tokens:  
    for w in i: 
        if w not in stop_words: 
          filtered_sentence.append(w) 
        
print("\n\nOriginal Sentence \n\n")
for i in word_tokens:
    print(" ".join(i)) 
```

```python
print("\n\nFiltered Sentence \n\n")
print(" ".join(filtered_sentence))
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()

Output=[stemmer.stem(word) for word in filtered_sentence]
print(Output)
```

```python
lemma_word = []
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
# v stands for verb, a stands for adjective and n stands for noun
for w in filtered_sentence:
    word1 = wordnet_lemmatizer.lemmatize(w, pos = "n")
    word2 = wordnet_lemmatizer.lemmatize(word1, pos = "v")
    word3 = wordnet_lemmatizer.lemmatize(word2, pos = ("a"))
    lemma_word.append(word3)
print(lemma_word)
```

```python
# import packages from scikit-learn
from sklearn.decomposition import LatentDirichletAllocation, NMF, TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = CountVectorizer(min_df=5, max_df=0.9, 
                             stop_words='english', lowercase=True, 
                             token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')
data_vectorized = vectorizer.fit_transform(lemma_word)
```

```python
# Build a Latent Dirichlet Allocation Model
lda_model = LatentDirichletAllocation(n_components= 10, max_iter=10, learning_method='online') # no of components = no. of topics
lda_Z = lda_model.fit_transform(data_vectorized)
print(lda_Z.shape)  # (NO_DOCUMENTS, NO_TOPICS)
 
# Build a Non-Negative Matrix Factorization Model
nmf_model = NMF(n_components= 10)
nmf_Z = nmf_model.fit_transform(data_vectorized)
print(nmf_Z.shape)  # (NO_DOCUMENTS, NO_TOPICS)
 
# Build a Latent Semantic Indexing Model
lsi_model = TruncatedSVD(n_components=10)
lsi_Z = lsi_model.fit_transform(data_vectorized)
print(lsi_Z.shape)  # (NO_DOCUMENTS, NO_TOPICS)
```

4. How many topics are there, what are they?



```python
# inspect the inferred topics
def print_topics(model, vectorizer, top_n=10):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-top_n - 1:-1]])
```

5. Which vectorization method performed better?

6. Compare the results with different topic classification models.


```python
print("LDA Model:")
print_topics(lda_model, vectorizer)
print("=" * 20)
```

```python
print("NMF Model:")
print_topics(nmf_model, vectorizer)
print("=" * 20)
```

```python
print("LSI Model:")
print_topics(lsi_model, vectorizer)
print("=" * 20)
```
