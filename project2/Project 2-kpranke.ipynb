# Project Two: Topic Classification (unsupervised classification)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
# set of stop words
stop_words = set(stopwords.words('english')) 
```

```python
# Load The Project Gutenberg eBook of Meditations, by Marcus Aurelius
import urllib
url = "https://www.gutenberg.org/files/2680/2680-0.txt"
file = urllib.request.urlopen(url)

text = []

for line in file:
    decoded_line = line.decode("utf-8")
    print(decoded_line)
    text.append(decoded_line)
```

```python
text
```

```python
def preprocess_review(text):
    #remove punctuation
    tokenized = word_tokenize(str(text)) # tokenize
    tokenized_alnum = [word for word in tokenized if word.isalnum()] #alphanum only
    tokens_lowercase = [token.lower() for token in tokenized_alnum] #change all to lowercase
    lemmatized = [lemmatizer.lemmatize(token) for token in tokens_lowercase] # lemmatize
    #add POS
    filtered_words = [word for word in lemmatized if word not in stopwords.words('english')] #remove stopwords
    return str(filtered_words)
```

```python
preprocessed_text = preprocess_review(text)
```

```python
print(preprocessed_text)
```

```python
preprocessed_text = [preprocessed_text]
```

```python
# import packages from scikit-learn
from sklearn.decomposition import LatentDirichletAllocation, NMF, TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = CountVectorizer(min_df=1, max_df=1.0, 
                             stop_words='english', lowercase=True, 
                             token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')
data_vectorized = vectorizer.fit_transform(preprocessed_text)
```

```python
# Build a Latent Dirichlet Allocation Model
lda_model = LatentDirichletAllocation(n_components= 10, max_iter=10, learning_method='online') # no of components = no. of topics
lda_Z = lda_model.fit_transform(data_vectorized)
print(lda_Z.shape)  # (NO_DOCUMENTS, NO_TOPICS)
 
# Build a Non-Negative Matrix Factorization Model
nmf_model = NMF(n_components= 10)
nmf_Z = nmf_model.fit_transform(data_vectorized)
print(nmf_Z.shape)  # (NO_DOCUMENTS, NO_TOPICS)
 
# Build a Latent Semantic Indexing Model
lsi_model = TruncatedSVD(n_components=10)
lsi_Z = lsi_model.fit_transform(data_vectorized)
print(lsi_Z.shape)  # (NO_DOCUMENTS, NO_TOPICS)
```

```python
# inspect the inferred topics
def print_topics(model, vectorizer, top_n=10):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-top_n - 1:-1]])
```

```python
print("LDA Model:")
print_topics(lda_model, vectorizer)
print("=" * 20)
 
```

```python
print("NMF Model:")
print_topics(nmf_model, vectorizer)
print("=" * 20)
```

```python
print("LSI Model:")
print_topics(lsi_model, vectorizer)
print("=" * 20)
```
