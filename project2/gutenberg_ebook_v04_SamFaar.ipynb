```python
# Provide url for "Plain Text UTF-8" of the ebook from https://www.gutenberg.org
url = "https://www.gutenberg.org/files/135/135-0.txt"

def read_gutenberg(url):
    import urllib
    file = urllib.request.urlopen(url)
    txt_file = open('guten_ebook.txt','w+')
    
    for line in file:
        decoded_line = line.decode("utf-8")
        txt_file.write(decoded_line)
    
    print('=== guten_ebook.txt IS GIVEN BELOW === \n')
    with open('guten_ebook.txt','r') as txt:
        for line in txt:
            print(line, end='')
```

# Pre-Processing, Vectorizing & Unsupervised Topic Classification

> Below we define a custom function that does all the modeling steps:
    
        1- Pre-Processing: removing stopwords, stemming, lemmatization using WordNetLemmatizer and SnowballStemmer from nltk and simple_preprocess from Gensim
        2- Vectorizing: TfidfVectorizer or CountVectorizer based on user input
        3- Unsupervised Topic Classification from sklearn (not gensim): LDA, NMF, or LSI based on user input
---
    
    
> Our functiuon's outputs are:
    
        - Nb of Docoments & Nb of Topics
        - Top n topics with top m words in each topic

```python
# function for Pre-processing, Vectorizing & Modeling
def topic_classifier(text, vectorizer_method, model, n_components=10, top_n=8):
    """
    >>> text: Plain Text UTF-8 format
    >>> vectorizer_method: One of two sklearn vectorizors: TfidfVectorizer, CountVectorizer
    >>> model: One of three sklearn topic classifiers: LDA, NMF, LSI
    >>> n_components: Number of topics for modeling
    >>> top_n: Number of words to see in each topic result
    """
    from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
    from sklearn.decomposition import LatentDirichletAllocation, NMF, TruncatedSVD
    
    
    # pre-process: removing stopwords, stemming, lemmatization
    import gensim
    from gensim.utils import simple_preprocess   # Convert a document into a list of lowercase tokens, ignoring tokens that are too short or too long.
    from gensim.parsing.preprocessing import STOPWORDS
    import nltk
    from nltk.stem import WordNetLemmatizer, SnowballStemmer
    stemmer = SnowballStemmer('english')
    def lemmatize_stemming(text):
        return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))  # pos = "v" => only lemmatize verbs and leave nouns 'n' and adjectives 'a' alone
    def preprocess(text):
        token_list = []
        for token in gensim.utils.simple_preprocess(text):
            if token not in gensim.parsing.preprocessing.STOPWORDS:
                token_list.append(lemmatize_stemming(token))
        return token_list
    
    
    # vectorizing
    if str(vectorizer_method).lower() == 'countvectorizer':
        print('=== Topic classifying using CountVectorizer ===\n')
        vectorizer = CountVectorizer(min_df=5,                                     # minimum reqd occurences of a word
                                     max_df=0.9,
                                     stop_words='english',                         # remove stop words
                                     lowercase=True,                               # convert all words to lowercase
                                     token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')     # num chars > 2
    else:
        print('=== Topic classifying using TfidfVectorizer ===\n')
        vectorizer = TfidfVectorizer(min_df=5,
                                     max_df=0.9,
                                     stop_words='english',
                                     lowercase=True,
                                     token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')
    # we will call our function "preprocess" for the fit_transform
    # We also need to "open" the text file and "read" it -> open(text).read()
    data_vectorized = vectorizer.fit_transform(preprocess(open(text).read()))
    
    
    # custom function for inspecting the topics result of our modeling
    def print_topics(model, vectorizer, top_n=top_n):
        for idx, topic in enumerate(model.components_):
            print("Topic %d:" % (idx))
            print([(vectorizer.get_feature_names()[i], topic[i].round(6))
                   for i in topic.argsort()[:-top_n - 1:-1]])
    
    
    # modeling (LDA, NMF, LSI)
    from sklearn.decomposition import LatentDirichletAllocation, NMF, TruncatedSVD
    if str(model).lower() == 'lda':
        # Build a Latent Dirichlet Allocation Model
        lda_model = LatentDirichletAllocation(n_components=n_components, max_iter=10, learning_method='online')    # n_components = no. of topics
        lda_fit_cnt_v = lda_model.fit_transform(data_vectorized)
        print(f'LDA Nb of Docs & Nb of Topics: {lda_fit_cnt_v.shape}\n')
        # print topics
        print(f'Top {n_components} LDA topics with top {top_n} words in each topic:\n')
        print_topics(lda_model, vectorizer)
    
    elif str(model).lower() == 'nmf':
        # Build a Non-Negative Matrix Factorization Model
        # with random init, the results can be different every time; 
        # More reliable results can be obtained if you initialise with SVD (Belford et al, 2018).
        nmf_model = NMF(n_components=n_components, init="nndsvd")   
        nmf_fit_cnt_v = nmf_model.fit_transform(data_vectorized)
        print(f'NMF Nb of Docs & Nb of Topics: {nmf_fit_cnt_v.shape}\n')
        # print topics
        print(f'Top {n_components} NMF topics with top {top_n} words in each topic:\n')
        print_topics(nmf_model, vectorizer)
    
    else:
        # Build a truncated SVD (aka LSI: Latent Semantic Indexing) Model
        lsi_model = TruncatedSVD(n_components=n_components)
        lsi_fit_cnt_v = lsi_model.fit_transform(data_vectorized)
        print(f'LSI Nb of Docs & Nb of Topics: {lsi_fit_cnt_v.shape}\n')
        # print topics
        print(f'Top {n_components} LSI topics with top {top_n} words in each topic:\n')
        print_topics(lsi_model, vectorizer)
```

# Evaluate Vectorization Methods & Topic Models

> In this section, we will evaluate the three topic modeling methods (LDA, NMF, LSI), and score them for the following methods to quantify the change:
    
        1- LDA, NMF, LSI using BOW (bag of words, i.e. the frequency of words)
        2- LDA, NMF, LSI using Tf-Idf (i.e. by removing words based on Tf-Idf value)
        3- Add bi/tri-grams to (2)

> The models will be scored usign **Coherence**, which scores a single topic by measuring the degree of semantic similarity between high scoring words in the topic.

> We will use **Gensim** library (instead of sklearn) since it has a great Topic Coherence Pipeline.

```python
import numpy as np
import glob

#Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

#spacy
import spacy
from nltk.corpus import stopwords
nlp = spacy.load("en_core_web_lg")
nlp.max_length = 4000000

import warnings
warnings.filterwarnings('ignore')
```

> We start by pre-processing using our custom "preprocess" function.

```python
# pre-process
import gensim
from gensim.utils import simple_preprocess  # Convert a document into a list of lowercase tokens, ignoring tokens that are too short or too long.
from gensim.parsing.preprocessing import STOPWORDS
import nltk
from nltk.stem import WordNetLemmatizer, SnowballStemmer

stemmer = SnowballStemmer('english')


def lemmatize_stemming(text):
    return stemmer.stem(
        WordNetLemmatizer().lemmatize(text, pos='v')
    )  # pos = "v" => only lemmatize verbs and leave nouns 'n' and adjectives 'a' alone


def preprocess(text):
    token_list = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS:
            token_list.append(lemmatize_stemming(token))
    return token_list
```

```python
token_list = preprocess(open('les_miser.txt').read())
token_list
```

> Next, we create a Dictionary and a Corpus, which are the two main inputs to the our topic models:

        - dictionary (word_id, word)
        - corpus (word_id, word_frequency)

## Bag of Words (BOW)

```python
# id2word dictionary
id2word = corpora.Dictionary([token_list])
print(type(id2word))
print()
print(id2word)

# Create Corpus
corpus = [id2word.doc2bow(text) for text in [token_list]]   # corpus is a list of one list
print('\n',corpus[0][0:100])
```

> The "id2word' (which has a python Dictionary data type) contains 13741 unique words, as we can see above.

> Gensim creates a unique id for each word in the document. The produced corpus shown above is a mapping of (word_id, word_frequency):

        - For example, (0, 6) above implies, word id 0 occurs 6 times and so on.

> Below, we create a custom function with model name as input, and topics and coherence score as outputs:

```python
def gensim_model(model,
                 corpus=corpus,
                 id2word=id2word,
                 num_topics=10,
                 chunksize=100):
    """
    >>> model: lda, nmf, or lsi
    """
    
    global lda_model, nmf_model, lsi_model

    if model.lower() == 'lda':
        lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                    id2word=id2word,
                                                    num_topics=num_topics,
                                                    chunksize=chunksize)
        
        print('=== weight*word for each topic ===\n')
        for idx, topic in lda_model.print_topics(num_words=8):
            print(f'Topic{idx}: {topic}')
            
        # Compute Coherence Score
        coherence_model_lda = CoherenceModel(model=lda_model, texts=[token_list], dictionary=id2word, coherence='c_v')
        coherence_lda = coherence_model_lda.get_coherence()
        print('\nCoherence Score: ', coherence_lda.round(4))
        
    elif model.lower() == 'nmf':
        from gensim.models.nmf import Nmf as GensimNmf
        nmf_model = GensimNmf(corpus=corpus,
                              id2word=id2word,
                              num_topics=num_topics,
                              chunksize=chunksize)
        
        print('=== weight*word for each topic ===\n')
        for idx, topic in nmf_model.print_topics(num_words=8):
            print(f'Topic{idx}: {topic}')
        
        # Compute Coherence Score
        coherence_model_nmf = CoherenceModel(model=nmf_model, texts=[token_list], dictionary=id2word, coherence='c_v')
        coherence_nmf = coherence_model_nmf.get_coherence()
        print('\nCoherence Score: ', coherence_nmf.round(4))
        
    else:
        lsi_model = gensim.models.lsimodel.LsiModel(
            corpus=corpus,
            id2word=id2word,
            num_topics=num_topics,
            chunksize=chunksize)
        
        print('=== weight*word for each topic ===\n')
        for idx, topic in lsi_model.print_topics(num_words=8):
            print(f'Topic{idx}: {topic}')
        
        # Compute Coherence Score
        coherence_model_lsi = CoherenceModel(model=lsi_model, texts=[token_list], dictionary=id2word, coherence='c_v')
        coherence_lsi = coherence_model_lsi.get_coherence()
        print('\nCoherence Score: ', coherence_lsi.round(4))
```

> For the next step, we define a fuction that runs all three models using bag of words vectorizer and stores the coherence values into our model_results dataframe.

```python
model_results = pd.DataFrame(columns = ['model', 'feature', 'c_v'])
```

```python
def gensim_models_eval(corpus=corpus,
                      id2word=id2word,
                      num_topics=10,
                      chunksize=100,
                      feature='BOW'):
    global model_results
    # LDA
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=id2word,
                                                num_topics=num_topics,
                                                chunksize=chunksize)
    # LDA Coherence Score
    coherence_model_lda = CoherenceModel(model=lda_model,
                                         texts=[token_list],
                                         dictionary=id2word,
                                         coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()

    # NMF
    from gensim.models.nmf import Nmf as GensimNmf
    nmf_model = GensimNmf(corpus=corpus,
                          id2word=id2word,
                          num_topics=num_topics,
                          chunksize=chunksize)
    # NMF Coherence Score
    coherence_model_nmf = CoherenceModel(model=nmf_model,
                                         texts=[token_list],
                                         dictionary=id2word,
                                         coherence='c_v')
    coherence_nmf = coherence_model_nmf.get_coherence()

    # LSI
    lsi_model = gensim.models.lsimodel.LsiModel(corpus=corpus,
                                                id2word=id2word,
                                                num_topics=num_topics,
                                                chunksize=chunksize)
    # LSI Coherence Score
    coherence_model_lsi = CoherenceModel(model=lsi_model,
                                         texts=[token_list],
                                         dictionary=id2word,
                                         coherence='c_v')
    coherence_lsi = coherence_model_lsi.get_coherence()
    
    model_results = model_results.append({'model' : 'LDA', 'feature' : feature, 'c_v' : coherence_lda}, ignore_index = True)
    model_results = model_results.append({'model' : 'NMF', 'feature' : feature, 'c_v' : coherence_nmf}, ignore_index = True) 
    model_results = model_results.append({'model' : 'LSI', 'feature' : feature, 'c_v' : coherence_lsi}, ignore_index = True) 

    return model_results
```

```python
gensim_models_eval()
```

## Bi-/Tri-Grams

```python
# Bigrams and Trigrams
data_words = [token_list]
bigram_phrases = gensim.models.Phrases(data_words,
                                       min_count=5,    # min nb of times that a bigram needs to happen
                                       threshold=100)  # nb of phrases found (higher means fewer phrases)

trigram_phrases = gensim.models.Phrases(bigram_phrases[data_words],
                                        threshold=100)

bigram = gensim.models.phrases.Phraser(bigram_phrases)
trigram = gensim.models.phrases.Phraser(trigram_phrases)

def make_bigrams(texts):
    return [bigram[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram[bigram[doc]] for doc in texts]

data_bigrams = make_bigrams(data_words)
data_bigrams_trigrams = make_trigrams(data_bigrams)
```

# LDA with BayesSearch Cross Validation

> For this section, we will focus on sklearn LDA model and TfidfVectorizer. We will optimize two hyper-parameters: number of topics, and learning decay which controls the learning rate.

> BayesSearch is based on **Bayesian optimization, which is a sequential model-based optimization (SMBO)** algorithm using the results from the previous iteration to decide the next hyperparameter value candidates. So, instead of blindly searching the hyperparameter space (like in Grid Search and Random Search), this method advocates the usage of intelligence to pick the next set of hyperparameters which will improve the model performance. We iteratively repeat this process until we converge to an optimum <a href="https://medium.com/analytics-vidhya/comparison-of-hyperparameter-tuning-algorithms-grid-search-random-search-bayesian-optimization-5326aaef1bd1">(LINK)</a>.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=5,                                     # minimum reqd occurences of a word 
                             max_df=0.9,
                             stop_words='english',                         # remove stop words
                             lowercase=True,                               # convert all words to lowercase
                             token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')     # num chars > 2

data_vectorized = vectorizer.fit_transform(preprocess(open('les_miser.txt').read()))
```

```python
import numpy as np
from skopt import BayesSearchCV
from sklearn.model_selection import StratifiedKFold

params = {
    "n_components": np.arange(6, 31, 4),   # number of topics
    'learning_decay': [0.4, 0.6, 0.8]      # controls the learning rate
}

# Init the Model
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation()

n_iter = 1
lda_model_bay_cv = BayesSearchCV(
    estimator=lda_model,
    search_spaces=params,
    n_iter=n_iter,
    n_jobs=8,
    random_state=123)

lda_model_fit = lda_model_bay_cv.fit(data_vectorized)
```

```python
# Best Model
best_lda_model = lda_model_bay_cv.best_estimator_

# Model Parameters
print("Best Model's Params:", lda_model_bay_cv.best_params_)

# Log Likelihood Score
print("\nBest Log Likelihood Score:", lda_model_bay_cv.best_score_)

# Perplexity
print("\nModel Perplexity:", best_lda_model.perplexity(data_vectorized))
```

> We can see that the best hyperparameters are 14 topics and 0.8 for learning decay.
